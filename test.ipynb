```python
import numpy as np
from matplotlib import pyplot as plt
from hoeffdingd import hoeffdingd

X = np.random.normal(0.0, 1.0, 1000)
Y = np.random.normal(0.0, 1.0, 1000)

plt.clf()
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X, Y, alpha=0.5)
plt.show()

print(hoeffdingd(X, Y))
```

```python
X = np.random.normal(0.0, 1.0, 1000)
Y = X

plt.clf()
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X, Y, alpha=0.5)
plt.show()

print(hoeffdingd(X, Y))
```

```python
X = np.random.normal(0.0, 1.0, 1000)
Y = np.random.normal(0.0, 0.2, 1000) + X

plt.clf()
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X, Y, alpha=0.5)
plt.show()

print(hoeffdingd(X, Y))
```

```python
X = np.random.normal(0.0, 1.0, 1000)
Y = np.random.normal(0.0, 1.0, 1000) + X**3

plt.clf()
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X, Y, alpha=0.5)
plt.show()

print(hoeffdingd(X, Y))
```
